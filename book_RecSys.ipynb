```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import torch
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/full_df.csv', sep='*', escapechar='\\', quoting=csv.QUOTE_NONE)
df.head()
```

```python
df.shape
```

```python
# проверяем длину описаний
df['annotation'].fillna("").apply(lambda x: len(x.split()))
df['annotation'].fillna("").apply(lambda x: len(x.split())).describe()
```

```python
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2') # модель эмбеддингов Sentence-BERT
```

```python
# получаем эмбеддинги для всех книг в датасете
corpus = df['annotation'].tolist()
corpus_embeddings = model.encode(corpus, convert_to_tensor=True, show_progress_bar=True)
```

```python
# функция рекомендаций (вычисляем топ похожих книг на основе косинусной близости)
def recommend_books_sbert(query, top_n=5):
    query_embedding = model.encode(query, convert_to_tensor=True)
    cosine_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]
    top_results = torch.topk(cosine_scores, k=top_n)

    results = []
    for score, idx in zip(top_results[0], top_results[1]):
        idx = int(idx)
        results.append({
            'название': df.iloc[idx]['title'],
            'автор': df.iloc[idx]['author'],
            'описание': df.iloc[idx]['annotation'],
            'сходство': score.item() # будем отображать значение сходства
        })

    return pd.DataFrame(results)
```

```python
# пример пользовательского запроса
query = "что-то атмосферное, мистическое и философское"
recommend_books_sbert(query)
```

```python
# пример пользовательского запроса
query = "нестрашная уютная детективная история "
recommend_books_sbert(query)
```

```python
from huggingface_hub import InferenceClient
import os

HF_TOKEN = ""
client = InferenceClient(model="mistralai/Mixtral-8x7B-Instruct-v0.1", token=HF_TOKEN) # используем Mixtral-8x7B от Mistral AI
```

```python
# функция создания промта для LLM
def explain_book_choice(query, book):
    prompt = f"""
Ты — помощник рекомендательной системы книг. Пользователь задал запрос: "{query}".

Вот описание подходящей книги:
Название: {book['название']}
Автор: {book['автор']}
Описание: {book['описание']}

Объясни, почему эта книга хорошо соответствует запросу. Ответь на русском языке, коротко и ясно.
"""

    response = client.text_generation(
        prompt=prompt,
        max_new_tokens=200,
        temperature=0.7,
        top_p=0.9
    )

    return response.strip()
```

```python
def recommend_books_sbert_with_explanations(query, top_n=5):
    query_embedding = model.encode(query, convert_to_tensor=True)
    cosine_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]
    top_results = torch.topk(cosine_scores, k=top_n)

    results = []
    for score, idx in zip(top_results[0], top_results[1]):
        idx = int(idx)
        book = {
            'название': df.iloc[idx]['title'],
            'автор': df.iloc[idx]['author'],
            'описание': df.iloc[idx]['annotation']
        }
        explanation = explain_book_choice(query, book)
        results.append({
            **book,
            'сходство': score.item(),
            'объяснение': explanation
        })

    return pd.DataFrame(results)
```

```python
query = "что-то атмосферное, мистическое и философское"
recommendations = recommend_books_sbert_with_explanations(query)
display(recommendations)
```
